The programs written below are the practice of various cases in which the "labels" argument can be used.

```r
# Packages used to run the programs below
# Installing and loading the libraries
install.packages(c("ggplot2", "gridExtra", "directlabels"))
library(ggplot2)
library(gridExtra)
library(directlabels)
```

```r
# Display all plots with size 15 X 9
options(repr.plot.width = 15, repr.plot.height = 9)
```

```r
# ggplot2 and gridExtra packages used

# Create own data frame named "df"
df <- data.frame(
  x = c(3, 1, 5),
  y = c(2, 4, 6), 
  label = c("a", "b", "c")
)
p <- ggplot(df, aes(x, y, label = label)) + 
  labs(x = NULL, y = NULL) + 
  theme(plot.title = element_text(size = 16, family = "serif"))

# Adding a point geom
point <- p + geom_point(size = 4, color = "#BB4F63") + ggtitle("Point")

# Adding a text geom
text <- p + geom_text(color = "#F08357", size = 5) + ggtitle("Text")

# Adding a bar geom
bar <- p + geom_bar(stat = "identity", fill = "#6AA1E2") + ggtitle("Bar")

# Adding a tile geom
tile <- p + geom_tile(fill = "#DF7FF6") + ggtitle("Raster")

# Adding a line geom
line <- p + geom_line(size = 1.1, color = "#BB4F63") + ggtitle("Line")

# Adding a area geom
area <- p + geom_area(fill = "#F08357") + ggtitle("Area")

# Adding a path geom
path <- p + geom_path(size = 1.1, colour = "#6AA1E2") + ggtitle("Path")

# Adding a polygon geom
polygon <- p + geom_polygon(fill = "#DF7FF6") + ggtitle("Polygon")

# Arranging all the above plots into 4 X n grid
grid.arrange(point, text, bar, tile, line, area, path, polygon, ncol = 4)
```

```r
# ggplot2 and gridExtra packages used

# Create own data frame named "df"
df <- data.frame(x = 1, y = 3:1, family = c("sans", "serif", "mono"), 
                 face = c("plain", "bold", "italic"))

my.plot <- ggplot(df, aes(x, y)) 
  
# Various text family
family <- my.plot + geom_text(aes(label = family, family = family))

# various text fontface
face <- my.plot + geom_text(aes(label = face, fontface = face))

grid.arrange(family, face, ncol = 2)
```

```r
# ggplot2 and gridExtra packages used

# Create own data frame named "df"
df <- data.frame(
  x = c(1, 1, 2, 2, 1.5), 
  y = c(1, 2, 1, 2, 1.5), 
  text = c("bottom-left", "bottom-right", "top-left", "top-right", "center")
)

p <- ggplot(df, aes(x, y))

not_just <- p + geom_text(aes(label = text))
just <- p + geom_text(aes(label = text), vjust = "inward", hjust = "inward")

grid.arrange(not_just, just, ncol = 2)
```

```r
# ggplot2 and gridExtra packages used

# Create own data frame named "df"
df <- data.frame(trt = c("a", "b", "c"), resp = c(1.2, 3.4, 2.5))
ggplot(df, aes(resp, trt)) + 
  geom_point() + 
  geom_text(aes(label = paste0("(", resp, ")")), nudge_y = -0.25) + 
  xlim(1, 3.6)
```

```r
# ggplot2 package used

# Text plot displaying displ vs hwy
ggplot(mpg, aes(displ, hwy)) + 
  geom_text(aes(label = model), check_overlap = T) + 
  xlim(1, 8)
```

```r
# ggplot2 package used

# Create own data frame named "label"
label <- data.frame(
  waiting = c(55, 80), 
  eruptions = c(2, 4.3), 
  label = c("peak one", "peak two")
)

# Create a plot with tile and label geom
ggplot(faithfuld, aes(waiting, eruptions)) + 
  geom_tile(aes(fill = density)) + 
  geom_label(data = label, aes(label = label))
```

```r
# ggplot2 and directlabels packages used

# Create scatterplot of displ vs hwy
ggplot(mpg, aes(displ, hwy, colour = class)) + 
  geom_point(size = 4) + 
  geom_dl(aes(label = class), method = "smart.grid")
```
